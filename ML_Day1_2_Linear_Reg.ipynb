```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
data = pd.read_csv("/content/Salary_Data.csv")
print(data)
```

```python
sns.heatmap(data.isnull())
#print(data.isnull())
```

```python
avg_salary = data['Salary'].mean()
data['Salary'].fillna(avg_salary, inplace=True)
```

```python
sns.heatmap(data.isnull())
```

Separating the x and y arguments

```python
x = data.iloc[:,:-1].values
y = data.iloc[:,-1].values
print(data.iloc[:,:1].values)
```

```python
plt.scatter(x,y, color='r')
plt.xlabel('Year of Exp.')
plt.ylabel('Salary')
```

After analysising this we can say that the preferrable model is 'Linear Regre
ssion'

```python
 from sklearn.linear_model import LinearRegression
 model = LinearRegression()
```

Feeding the data

```python
model.fit(x,y)
```

Plotting the prediction line

```python
plt.scatter(x,y, color='r')
plt.plot(x, model.predict(x), color = 'b')
plt.xlabel('Year of Exp.')
plt.ylabel('Salary')
```

Predicting a output

```python
input_exp_year = [[3.5],[2.5]]
pridicted_salary = model.predict(input_exp_year)
print(pridicted_salary)
```

PLotting the predicted output

```python
plt.scatter(x,y, color='r')
plt.scatter(input_exp_year, pridicted_salary,color = 'g')
plt.plot(x, model.predict(x), color = 'b')
plt.xlabel('Year of Exp.')
plt.ylabel('Salary')
```
